# Номер1

$x′=kx$

```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy.integrate import odeint
```

```python
def dxdt(x, t): 
    k = 0.42
    return k * x
```

```python
t = np.linspace(0, 10, 100)
```

```python
solve = odeint(dxdt, y0=1, t=t) 
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$x' = kx$")
plt.grid(True)
plt.legend()
plt.show()
```

# Номер 2

$x'=k(x)x , k(x)=a-bx$

```python
def k(x): 
    a=1
    b=2
    return a-b*x
def dxdt(x, t): 
    return k(x)*x
```

# Номер 4

$x'=x-g(x)x-c$

```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy.integrate import odeint
```

```python
def g(x):
    a=1
    b=2
    return a-b*x
def dxdt(x, t):
    c=1
    return x * g(x) * x - c
```

```python
t = np.linspace(0, 10, 100)
```

```python
solve = odeint(dxdt, y0=1, t=t) 
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(t, solve, label = r"$x' = x-g(x)*x-c$")
plt.grid(True)
plt.legend()
plt.show()
```
